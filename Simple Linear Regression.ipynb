# Simple linear regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
X = df["cgpa"]
```

```python
y = df["package"]
```

```python
coefficients = np.polyfit(X, y, 1)
line = coefficients[0] * np.array(X) + coefficients[1]
plt.scatter(X,y)
plt.plot(X, line, color='red', label='Linear Regression Line')
plt.xlabel("CGPA")
plt.ylabel("Salary in MPA")
plt.title("Salary vs CGPA")
```
